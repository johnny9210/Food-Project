```python
from gensim.models import Word2Vec
from konlpy.tag import *
from pandas import Series, DataFrame
import pandas as pd
import sqlite3
import nltk
from nltk.corpus import stopwords
from gensim import corpora
import pyLDAvis.gensim
```

```python
con =sqlite3.connect('insta_comment.db')
cur=con.cursor()
cur.execute("select comments from food_table where keyword_food='냉동삼겹살'")
a=cur.fetchall()
print(a)
```

```python
comment=[]
for idx,val in enumerate(a): 
    for idx2, val2 in enumerate(val):
        comment.append(val2) 
```

```python
comment_df=pd.DataFrame({'comment_document':comment})
```

```python
comment_df
```

```python
tokenized_doc = comment_df['comment_document'].apply(lambda x: x.split())
```

```python
print(type(tokenized_doc))
print(tokenized_doc[1])
```

```python
dictionary = corpora.Dictionary(tokenized_doc)
print(dictionary)
corpus = [dictionary.doc2bow(text) for text in tokenized_doc]
print(corpus[20]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
```

```python
print(corpus[20])
```

```python
print(dictionary[294])
```

```python
NUM_TOPICS = '설정해야됨' #20개의 토픽, k=20
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```
