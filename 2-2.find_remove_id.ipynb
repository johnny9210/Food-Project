```python
import pandas as pd
import sqlite3
```

## 사용자별 개시글 갯수

```python
dbname = db위치 입력
sql = 'SELECT * FROM {}'.format( 테이블명 입력 )



colnames = ['idx_text' , 'user_id' , 'date' , 'comments' ,'like' ,'keyword_food']

with sqlite3.connect(dbname) as connection:
    cur = connection.cursor()
    raw_data = cur.execute(sql).fetchall()
    raw_data = pd.DataFrame( raw_data , columns = colnames )

    
    
unique_id = list(raw_data['uesr_id'].unique())

comment_id = []
for id_ in unique_id:
    comment_id.append((id_, len( raw_data.loc[raw_data['uesr_id'] == id_] )))
```

```python
df = pd.DataFrame(comment_id , columns = ['id', 'count'])
df.sort_values('count' , ascending = False ).head(1000)

# df_sort = df.sort_values('count' , ascending = False )
# df_sort.to_csv('fake_id.csv' , mode = 'w' ) # 파일저장 하기
```
